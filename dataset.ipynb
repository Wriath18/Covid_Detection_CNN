```python
import os
```

```python
Target_dir = 'Datasets/Covid_Positive'

if not os.path.exists(Target_dir):
    os.mkdir(Target_dir)
    print("Folder created")
```

Moving the positive images to the folder

```python
count = 0
for (i,row) in data.iterrows():
    if row['finding'] == "Pneumonia/Viral/COVID-19" and row['view'] == "PA":
        count += 1
        filename = row['filename']
        image_path = os.path.join("./images/", filename)
        image_copy = os.path.join(Target_dir, filename)
        shutil.copy2(image_path, image_copy)
        print(f"Moving image : {count}")

print(count)
```

```python
import random
kaggle_images = 'chest_xray/train/NORMAL'
```

```python
target_normal = "Datasets/Normal"
image_names = os.listdir(kaggle_images)
random.shuffle(image_names)
```
